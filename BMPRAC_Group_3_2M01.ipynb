### ✅ **Step 1: Import Libraries and Define NHANES Cycles**

This step loads the required Python libraries for data analysis, statistical testing, and visualisation.  
It also defines the NHANES survey years and corresponding cycles (E to J) to be merged for analysis.

```python
# Import relevant libraries for analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
from scipy.stats import shapiro
import statsmodels.api as sm

# Define years & NHANES cycles for merging data
years = ["2007", "2009", "2011", "2013", "2015","2017"]
cycles = ["E", "F", "G", "H", "I","J"]
```

### ✅ **Step 2: Load and Extract NHANES Variables**

In this step, data is imported from NHANES for the selected survey years (2007–2018).  
The following variables are extracted from their respective datasets:
- **Demographics**: Gender and Age (`RIAGENDR`, `RIDAGEYR`)
- **BMI**: Body Mass Index (`BMXBMI`)
- **Liver Enzymes**: ALT, AST, GGT (`LBXSATSI`, `LBXSASSI`, `LBXSGTSI`)
- **Physical Activity**: Moderate work activity (`PAQ620`)
- **Hypertension History**: Blood pressure diagnosis (`BPQ020`)
- **Smoking Status**: Lifetime and current smoking (`SMQ020`, `SMQ040`)
- **Alcohol Use**: Average consumption (`ALQ130`)

Each variable is filtered and combined across six NHANES cycles, then stored in a corresponding dataframe for further cleaning and analysis.

```python
# Helper function to read and label each cycle
def read_and_tag(urls, selected_cols):
    dfs = []
    for y, c, url in zip(years, cycles, urls):
        try:
            df_temp = pd.read_sas(url)
            # Select only the columns that exist in the current dataframe
            cols_to_select = [col for col in selected_cols if col in df_temp.columns]
            df_temp = df_temp[cols_to_select + ["SEQN"]] # Always include SEQN for merging
            df_temp["Cycle"] = c
            dfs.append(df_temp)
        except Exception as e:
            print(f"Error reading {url}: {e}")
            # Optionally, you can log or handle the error differently
            continue # Skip this file if there's an error

    # Ensure all dataframes have the same columns before concatenating by filling missing ones with NaN
    all_cols = list(set([col for df in dfs for col in df.columns]))
    for i in range(len(dfs)):
        for col in all_cols:
            if col not in dfs[i].columns:
                dfs[i][col] = np.nan

    return pd.concat(dfs, ignore_index=True)

# Demographics (Age, Gender)
demo_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DEMO_{c}.xpt" for y, c in zip(years, cycles)]
df_demo = read_and_tag(demo_urls, ["RIAGENDR", "RIDAGEYR"])

# BMI
bmi_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BMX_{c}.xpt" for y, c in zip(years, cycles)]
df_bmi = read_and_tag(bmi_urls, ["BMXBMI"])

# Liver enzymes (ALT, AST, GGT) levels
liver_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BIOPRO_{c}.xpt" for y, c in zip(years, cycles)]
df_liver = read_and_tag(liver_urls, ["LBXSATSI", "LBXSASSI", "LBXSGTSI"])

# Physical activity
physical_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/PAQ_{c}.xpt" for y, c in zip(years, cycles)]
df_physical = read_and_tag(physical_urls, ["PAQ620"])

# Hypertension history
hypertension_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BPQ_{c}.xpt" for y, c in zip(years, cycles)]
df_hypertension = read_and_tag(hypertension_urls, ["BPQ020"])

# Smoking
smoking_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/SMQ_{c}.xpt" for y, c in zip(years, cycles)]
df_smoking = read_and_tag(smoking_urls, ["SMQ020", "SMQ040"])

# Alcohol use
alcohol_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/ALQ_{c}.xpt" for y, c in zip(years, cycles)]
df_drinking = read_and_tag(alcohol_urls, ["ALQ130"])
```

### ✅ **Step 3: Merge NHANES Data and Filter by Age**

This step merges all selected NHANES datasets using the participant ID (`SEQN`) and filters the data to include only adults aged 20 years and above, as required by the research question.

```python
# Merge all datasets on ["SEQN", "Cycle"]
df = df_demo.merge(df_bmi, on=["SEQN", "Cycle"], how="left") \
            .merge(df_liver, on=["SEQN", "Cycle"], how="left") \
            .merge(df_physical, on=["SEQN", "Cycle"], how="left") \
            .merge(df_hypertension, on=["SEQN", "Cycle"], how="left") \
            .merge(df_smoking, on=["SEQN", "Cycle"], how="left") \
            .merge(df_drinking, on=["SEQN", "Cycle"], how="left")

# Filter data for participants aged 20 and above
df = df[df["RIDAGEYR"] >= 20]

# Print shape after filtering
print("Shape after merging and age filtering:", df.shape)

# Print columns
print(df.columns.tolist())

# Map cycle letters to survey years
cycle_to_year = {
    "E": 2007,
    "F": 2009,
    "G": 2011,
    "H": 2013,
    "I": 2015,
    "J": 2017
}

# Add Year column
df["Year"] = df["Cycle"].map(cycle_to_year)
```

### ✅ **Step 5: Count Invalid Responses and Categorise Variables**

This step:
- Counts the number of participants who responded with **7/777 ("Refused")** or **9/999 ("Don't know")** for key categorical variables (e.g. alcohol use, smoking, physical activity).
- Replaces those invalid responses with `NaN` to prepare for cleaning.
- Categorises raw NHANES codes into more interpretable labels:
  - **Smoking** (e.g. Never smoked, Regular smoker)
  - **Alcohol Use** (e.g. Light drinker, Heavy drinker)
  - **Gender** (Male / Female)
  - **Physical Activity** (Yes / No)
  - **Hypertension History** (Yes / No / Borderline)

```python
# Count how many responded 7 ("Refused") or 9 ("Don't know")
coded_vars = ["SMQ020", "SMQ040", "PAQ620", "BPQ020"]

print("Counts of 7 = Refused and 9 = Don't know (before replacing with NaN):\n")
for col in coded_vars:
    if col in df.columns:
        print(f"{col} -- 7: {df[col].value_counts().get(7, 0)}, 9: {df[col].value_counts().get(9, 0)}")

# Count how many responded 777 ("Refused") or 999 ("Don't know")
coded_777_999_vars = ["ALQ130"]

print("Counts of 777 = Refused and 999 = Don't know (before replacing with NaN):\n")
for col in coded_777_999_vars:
    if col in df.columns:
        print(f"{col} -- 777: {df[col].value_counts().get(777, 0)}, 999: {df[col].value_counts().get(999, 0)}")

# Check unique values before replacement
print("\nUnique values in SMQ020 before replacement:", df["SMQ020"].unique())


# Categorise Smoking Status

# ----------------------------
# SMQ020: Ever smoked ≥ 100 cigarettes?
#     1 = Yes (Smoker)
#     2 = No (Non-smoker)
# ----------------------------
# SMQ040: Do you now smoke?
#     1 = Every day
#     2 = Some days
#     3 = Not at all
# ----------------------------

def classify_smoking(row):
    if pd.isna(row["SMQ020"]):
        return np.nan
    if row["SMQ020"] == 2:
        return "Never smoked"
    elif row["SMQ020"] == 1:
        if pd.isna(row["SMQ040"]):
            return np.nan
        if row["SMQ040"] == 1:
            return "Regular smoker"
        elif row["SMQ040"] == 2:
            return "Occasional smoker"
        elif row["SMQ040"] == 3:
            return "Reformed non-smoker"
    return np.nan

# Apply categorisation and clean raw codes
df["Smoking"] = df.apply(classify_smoking, axis=1)
df["SMQ020"] = df["SMQ020"].replace({7: np.nan, 9: np.nan})
df["SMQ040"] = df["SMQ040"].replace({7: np.nan, 9: np.nan})
print (df["Smoking"].value_counts())


# Categorise alcohol use
df["ALQ130"] = df["ALQ130"].replace({777: np.nan, 999: np.nan})

df["AlcoholUse"] = df["ALQ130"].map(
    lambda x: "Light drinker" if 1 <= x <= 13
    else "Heavy drinker" if x == 15
    else np.nan
)


# Categorise Gender
df["Gender"] = df["RIAGENDR"].map({1: "Male", 2: "Female"})


# Categorise Physical Activity
df["PAQ620"] = df["PAQ620"].replace({7: np.nan, 9: np.nan})
df["Activity"] = df["PAQ620"].map({1: "Yes", 2: "No"})


# Categorise Hypertension History
df["BPQ020"] = df["BPQ020"].replace({7: np.nan, 9: np.nan})
df["Hypertension"] = df["BPQ020"].map({
    1: "Yes",
    2: "No",
    3: "Borderline"
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import numpy as np

# ALT after cleaning
alt = df["LBXSATSI"]

# Summary statistics
mean = alt.mean()
median = alt.median()
q1 = alt.quantile(0.25)
q3 = alt.quantile(0.75)
min_val = alt.min()
max_val = alt.max()
std_dev = alt.std()
skew = alt.skew()

print("📊 Descriptive Statistics for ALT (LBXSATSI):")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"25th Percentile (Q1): {q1:.2f}")
print(f"75th Percentile (Q3): {q3:.2f}")
print(f"Minimum: {min_val:.2f}")
print(f"Maximum: {max_val:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skew:.2f}")

# Shapiro-Wilk test (with caution)
shapiro_stat, shapiro_p = stats.shapiro(alt)
print("\n⚠️ Shapiro-Wilk Test Warning:")
print("The sample size is large (n = 5,028). The test is highly sensitive and may detect minor deviations.")
print(f"Shapiro-Wilk Test: W = {shapiro_stat:.4f}, p = {shapiro_p:.4f}")

# Interpretation of p-value
if shapiro_p < 0.05:
    print("Interpretation: p < 0.05 → The data is likely **not normally distributed**.")
else:
    print("Interpretation: p ≥ 0.05 → The data **may be normally distributed**.")

# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(alt, bins=100, kde=True)
plt.title("Histogram of ALT (LBXSATSI)")
plt.xlabel("ALT (U/L)")
plt.ylabel("Frequency")
plt.show()

# Q-Q Plot
plt.figure(figsize=(6, 6))
stats.probplot(alt, dist="norm", plot=plt)
plt.title("Q-Q Plot of ALT (LBXSATSI)")
plt.show()
```

## **Section 7: Annual Trends**

This section presents yearly trends from 2007 to 2018, including the median levels of liver enzymes (ALT, AST, GGT) and BMI across NHANES cycles. It also highlights the annual percentage of participants with abnormal liver enzyme values based on established clinical thresholds. Sample sizes for each cycle are annotated to provide context for the observed patterns.

**⚠️ Liver Enzyme Thresholds for Abnormality**

The following thresholds were used to classify participants as having abnormal liver enzyme levels based on clinical guidelines and gender-specific reference ranges:

- ALT (Alanine Aminotransferase): *(Lee et al., 2010)*

≥ 33 U/L for males

≥ 25 U/L for females

- AST (Aspartate Aminotransferase): *(Fletcher & Sissons, 2025)*

≥ 40 U/L for all participants

- GGT (Gamma-Glutamyl Transferase): *(van Beek et al., 2014)*

≥ 51 U/L for males

≥ 33 U/L for females

Participants exceeding these thresholds were flagged as having elevated or abnormal liver enzyme levels, which may indicate potential liver dysfunction or stress.

```python
# Map NHANES Cycle Letters to Survey Years
cycle_to_year = {
    "E": 2007,
    "F": 2009,
    "G": 2011,
    "H": 2013,
    "I": 2015,
    "J": 2017
}

# Apply the mapping to create a 'Year' column
df["Year"] = df["Cycle"].map(cycle_to_year)
```

```python
import matplotlib.pyplot as plt

# Group by Year and calculate medians
trend_medians = df.groupby("Year")[["LBXSATSI", "LBXSASSI", "LBXSGTSI", "BMXBMI"]].median()

# Get participant counts per year
sample_sizes = df["Year"].value_counts().sort_index()

# Variable mapping for cleaner labels
var_labels = {
    "LBXSATSI": "ALT",
    "LBXSASSI": "AST",
    "LBXSGTSI": "GGT",
    "BMXBMI": "BMI"
}

# Loop and plot each variable separately
for var, label in var_labels.items():
    plt.figure(figsize=(7, 4))
    plt.plot(trend_medians.index, trend_medians[var], marker='o', color='steelblue')

    # Annotate with sample size
    for year in trend_medians.index:
        n = sample_sizes.get(year, 0)
        plt.annotate(f'n={n}', (year, trend_medians.loc[year][var]),
                     textcoords="offset points", xytext=(0, 8), ha='center', fontsize=9, color='gray')

    plt.title(f"Annual Median {label} with Sample Sizes")
    plt.xlabel("Survey Year")
    plt.ylabel(f"Median {label}")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
# Determine threholds for abnormal enzyme levels
# Abnormal ALT: Gender-specific thresholds
df["Abnormal_ALT"] = df.apply(
    lambda row: 1 if (
        (row["Gender"] == "Male" and row["LBXSATSI"] >= 33) or
        (row["Gender"] == "Female" and row["LBXSATSI"] >= 25)
    ) else 0,
    axis=1
)

# Abnormal AST: threshold ≥ 40 U/L
df["Abnormal_AST"] = df["LBXSASSI"].apply(lambda x: 1 if x >= 40 else 0)

# Abnormal GGT: Gender-specific thresholds
df["Abnormal_GGT"] = df.apply(
    lambda row: 1 if (
        (row["Gender"] == "Male" and row["LBXSGTSI"] >= 51) or
        (row["Gender"] == "Female" and row["LBXSGTSI"] >= 33)
    ) else 0,
    axis=1
)


# Obtain counts of abnormal enzyme levels
# Abnormal ALT count by gender
alt_counts = df[df["Abnormal_ALT"] == 1]["Gender"].value_counts()
print("🔬 Abnormal ALT Counts by Gender:")
print(alt_counts)
print()

# Abnormal AST total count (not gender-specific)
ast_count = df["Abnormal_AST"].sum()
print(f"🔬 Total Abnormal AST Count (≥ 40 U/L): {ast_count}")
print()

# Abnormal GGT count by gender
ggt_counts = df[df["Abnormal_GGT"] == 1]["Gender"].value_counts()
print("🔬 Abnormal GGT Counts by Gender:")
print(ggt_counts)
```

```python
# Step 1: Import prescription medication data (RXQ_RX)
rx_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/RXQ_RX_{c}.xpt"
           for y, c in zip(years, cycles)]

df_rx = pd.concat([pd.read_sas(url) for url in rx_urls])
df_rx = df_rx[["SEQN", "RXDDRUG", "RXDDRGID"]]  # RXDDRUG = drug name, RXDDRGID = drug ID

# Add cycle information
repeated_cycles = np.tile(cycles, int(np.ceil(len(df_rx) / len(cycles))))[:len(df_rx)]
df_rx["Cycle"] = repeated_cycles

# Decode bytes to strings and lowercase
df_rx["RXDDRUG_lower"] = df_rx["RXDDRUG"].apply(lambda x: x.decode("utf-8").lower() if isinstance(x, bytes) else str(x).lower())

# Define keywords related to liver disease treatment
liver_meds = [
    "ursodiol", "ursodeoxycholic", "lactulose", "ribavirin", "interferon", "entecavir",
    "tenofovir", "lamivudine", "adefovir", "simeprevir", "sofosbuvir"
]

# Flag if participant is taking any liver-related med
df_rx["LiverMed"] = df_rx["RXDDRUG_lower"].apply(
    lambda x: 1 if any(med in x for med in liver_meds) else 0
)

# Use temp copy of cleaned df
df_temp = df.copy()

# Merge with prescription data
df_temp = df_temp.merge(df_rx[["SEQN", "Cycle", "LiverMed"]], on=["SEQN", "Cycle"], how="left")

# Determine if participant took any liver-related meds
df_temp["On_Liver_Med"] = df_temp.groupby("SEQN")["LiverMed"].transform("max")

# Calculate percentage
n_total = df_temp["On_Liver_Med"].notna().sum()
n_on_meds = df_temp["On_Liver_Med"].sum()
percent_on_meds = (n_on_meds / n_total) * 100

# Display result
print(f"Taking liver-related meds: {int(n_on_meds)} out of {int(n_total)} participants")
print(f"Percentage: {percent_on_meds:.2f}%")

# Plot annual trend of participants on liver-related medications
import matplotlib.pyplot as plt

# Step: Add Year column (if not already done)
cycle_to_year = {"E": 2007, "F": 2009, "G": 2011, "H": 2013, "I": 2015, "J": 2017}
df_temp["Year"] = df_temp["Cycle"].map(cycle_to_year)

# Step: Group by year and calculate percentages
yearly_med_counts = df_temp.groupby("Year")["On_Liver_Med"].agg(["sum", "count"])
yearly_med_counts["Percent"] = (yearly_med_counts["sum"] / yearly_med_counts["count"]) * 100

# Step: Plot with sample size annotations
plt.figure(figsize=(8, 5))
plt.plot(yearly_med_counts.index, yearly_med_counts["Percent"], marker='o', color='tab:blue', label="Liver Meds (%)")

# Annotate each point with sample size (n)
for year in yearly_med_counts.index:
    n = yearly_med_counts.loc[year, "count"]
    percent = yearly_med_counts.loc[year, "Percent"]
    plt.annotate(f'n={n}', (year, percent),
                 textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9, color='gray')

plt.title("Annual Percentage of Participants Taking Liver-Related Medications")
plt.xlabel("Survey Year")
plt.ylabel("Percentage (%)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
